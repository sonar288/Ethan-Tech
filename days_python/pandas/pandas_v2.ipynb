```python
def vowel_counter(word):
    ctr=0
    vowels=['a', 'e', 'i', 'o', 'u']
    for char in word:
        if char in vowels:
            ctr+=1
    return ctr
```

```python
df.BusinessTravel.apply(lambda x : vowel_counter(x))
```

```python
df.apply(lambda x:vowel_counter(x.BusinessTravel), axis=1)
```

```python
#### pivot table
```

```python
pivot1 = pd.pivot_table(
    df,
    values="EmployeeNumber",   # any column (used just for counting)
    index="Department",
    columns="Attrition",
    aggfunc="count"
)
pivot1
```

```python
pivot2 = pd.pivot_table(
    df,
    values="MonthlyIncome",
    index="JobRole",
    aggfunc="mean"
)
pivot2
```

```python
pivot3 = pd.pivot_table(
    df,
    values="EmployeeNumber",
    index="EducationField",
    columns="Attrition",
    aggfunc="count",
    fill_value=0
)

# Calculate % attrition
pivot3["AttritionRate"] = pivot3["Yes"] / (pivot3["Yes"] + pivot3["No"])
pivot3
```

```python
pivot4 = pd.pivot_table(
    df,
    values=["Age", "MonthlyIncome"],
    index="Department",
    aggfunc={"Age": "mean", "MonthlyIncome": ["mean", "max"]}
)
pivot4
```
